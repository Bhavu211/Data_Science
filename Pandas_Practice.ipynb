**Adding statistical computation on groupby**

```python
# prompt: Adding statistical computation on groupby

import pandas as pd
import numpy as np

# Sample DataFrame (replace with your actual data)
data = {'Category': ['A', 'A', 'B', 'B', 'A', 'C', 'C', 'B'],
        'Value': [10, 15, 20, 25, 12, 30, 35, 22]}
df = pd.DataFrame(data)

# Group data by 'Category' and compute various statistics
grouped = df.groupby('Category')['Value']
result = grouped.agg(['sum', 'mean', 'median', 'std', 'min', 'max', 'count'])
result
```

**Using Filter Function with Groupby**

```python
# prompt: Using Filter Function with Groupby

# Sample DataFrame (replace with your actual data)
data = {'Category': ['A', 'A', 'B', 'B', 'A', 'C', 'C', 'B'],
        'Value': [10, 15, 20, 25, 12, 30, 35, 22]}
df = pd.DataFrame(data)

# Group data by 'Category' and filter groups where the sum of 'Value' is greater than 50
def filter_func(x):
  return x['Value'].sum() > 50

filtered_df = df.groupby('Category').filter(filter_func)

filtered_df
```

**3. Working with csv files and basic data analysis using Pandas**

```python
# prompt: Working with csv files and basic data analysis using Pandas

import pandas as pd

# Assuming the CSV file is named 'your_file.csv' and is in the same directory as the notebook
# Replace 'your_file.csv' with the actual file name if different
file_path = '/content/sample_data/california_housing_test.csv'

try:
    # a) Reading csv
    df = pd.read_csv(file_path)
    print("Successfully read the CSV file.")

    # b) Info function
    print("\nDataFrame Info:")
    df.info()

    # c) isnull() Function
    print("\nNull values count per column:")
    print(df.isnull().sum())

    # d) Quantile function (example: 25th percentile)
    print("\n25th Percentile of numerical columns:")
    print(df.quantile(0.25))

    # e) Copy function
    df_copy = df.copy()
    print("\nCreated a copy of the DataFrame.")

    # f) Value Counts function (example: for a column named 'Category')
    if 'Category' in df.columns:
      print("\nValue counts for 'Category' column:")
      print(df['Category'].value_counts())

    # g) Unique and Nunique function (example: for a column named 'Category')
    if 'Category' in df.columns:
      print("\nUnique values in 'Category' column:")
      print(df['Category'].unique())
      print("\nNumber of unique values in 'Category' column:")
      print(df['Category'].nunique())

    # h) dropna() function (example: drop rows with any missing values)
    df_dropped = df.dropna()
    print("\nDataFrame after dropping rows with missing values:")
    print(df_dropped)

    # i) fillna() function (example: fill missing values with 0)
    df_filled = df.fillna(0)
    print("\nDataFrame after filling missing values with 0:")
    print(df_filled)

    # j) sample function (example: get a random sample of 5 rows)
    df_sample = df.sample(n=5)  # Or use frac=0.1 for a fraction of the data
    print("\nRandom sample of the DataFrame:")
    print(df_sample)

    # k) to_csv() function (example: save the modified DataFrame to a new CSV file)
    df_filled.to_csv("modified_data.csv", index=False) # Set index=False to avoid saving the index
    print("\nModified DataFrame saved to 'modified_data.csv'")

except FileNotFoundError:
    print(f"Error: File '{file_path}' not found. Please ensure the file exists and the path is correct.")
except pd.errors.EmptyDataError:
    print(f"Error: File '{file_path}' is empty.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

**j) sample function**

```python
# prompt: sample function

import pandas as pd

def sample_dataframe(df, n=5, replace=False, random_state=None):
    """
    Generates a random sample from a Pandas DataFrame.

    Args:
        df: The input DataFrame.
        n: The number of rows to sample. Default is 5.
        replace: Whether to sample with replacement. Default is False.
        random_state: Seed for the random number generator. Default is None.

    Returns:
        A new DataFrame containing the sampled rows or None if the DataFrame is empty.
    """
    if df.empty:
        print("DataFrame is empty.")
        return None

    try:
        sample_df = df.sample(n=n, replace=replace, random_state=random_state)
        return sample_df
    except ValueError as e:
        print(f"Error generating sample: {e}")
        return None
```
