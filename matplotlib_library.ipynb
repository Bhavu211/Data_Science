**Sin and cos graph**

```python
x=np.linspace(0,10)
#Defining Wave
y=np.sin(x)
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10)
#Defining Wave
y=np.sin(x)
z=np.cos(x)
#Plotting the Wave
plt.plot(x,y,label='Sine', marker='*')
plt.plot(x,z,label='Cosine', linestyle='--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Sine and Cosine')
plt.legend()
plt.ylabel('y axis')
plt.title('Sine and Cosine')
plt.legend()
```

**Pie Charts**

```python
# prompt: Pie Charts

# **Pie Charts**
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python
# prompt: Pie Charts

# **Pie Charts**
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode=[0,0,0,.1]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python

# **Pie Charts**
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode=[0,0,0,.1]
plt.pie(sizes, labels=labels, shadow=True, autopct='%1.1f%%', startangle=90, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python

# **Pie Charts**
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode=[0,0,0,.1]
plt.pie(sizes, labels=labels,wedgeprops={'edgecolor':'black','linewidth':2, 'antialiased':False}, autopct='%1.1f%%', startangle=90, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart Example')
plt.show()
```

```python
plt.subplots_adjust(hspace= .4,wspace= .4)
for i in range(1,7):
  plt.subplot(2,3,i)
  plt.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')
plt.show()
```

```python
x=np.linspace(0,10,50)
y=np.sin(x)
for i in range(1,4):
  fig,ax=plt.subplots(1,3,figsize=(12,4))
  ax[0].plot(x,y)
  ax[1].plot(x,y**2)
  ax[2].plot(x,y**3)
  ax[0].set_title('Sine')
  ax[1].set_title('Square')
  ax[2].set_title('Cube')
  plt.tight_layout()
  plt.show()
```
